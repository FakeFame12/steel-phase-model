```python
import matplotlib.pyplot as plt
from pycalphad import Database, variables as v
from scheil import simulate_scheil_solidification
```

```python
# Setup the simulation parameters
dbf = Database('FeMnC_11Dju.tdb')
comps = ['FE', 'C', 'MN', 'VA']
phases = sorted(dbf.phases.keys())

liquid_phase_name = 'LIQUID'
initial_composition = {v.X('FE'): 0.910198, v.X('MN'): 0.080611}
start_temperature = 5000
```

```python
# Perform the simulation
sol_res = simulate_scheil_solidification(dbf, comps, phases, initial_composition, start_temperature, step_temperature=10.0, adaptive=True)
```

```python
# Plot the result
for phase_name, amounts in sol_res.cum_phase_amounts.items():
    if any(amount > 1e-3 for amount in amounts):
        plt.plot(sol_res.temperatures, amounts, label=phase_name)
plt.plot(sol_res.temperatures, sol_res.fraction_liquid, label='LIQUID')
plt.ylabel('Phase Fraction')
plt.xlabel('Temperature (K)')
plt.title('Fe-C-Mn Scheil simulation, phase fractions')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()

plt.plot(sol_res.temperatures, sol_res.fraction_solid)
plt.title("Fraction of Solid vs. Temperature(K)")
plt.xlabel("Temperature (K)")
plt.ylabel("Fraction of Solid")
plt.show()
```
